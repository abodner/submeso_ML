```python
import numpy as np
import matplotlib.pyplot as plt
import os
import xarray as xr
import seaborn as sns
```

## agulhas: load data 

```python
CASE_NAME = 'agulhas/'
BASE = '/scratch/ab10313/submeso_ML_data/'
PATH = BASE+CASE_NAME

# Load all datasets
PATH_PP = PATH+'preprocessed_data/'

# submesoscale fields
UsBs_MLD_lowres = xr.open_dataarray(PATH_PP+'UsBs_MLD_lowres.nc').isel(lon=slice(0,-3))
VsBs_MLD_lowres = xr.open_dataarray(PATH_PP+'VsBs_MLD_lowres.nc').isel(lon=slice(0,-3))
WsBs_MLD_lowres = xr.open_dataarray(PATH_PP+'WsBs_MLD_lowres.nc').isel(lon=slice(0,-3))

# buoyancy gradients
Bm_x_MLD_lowres = xr.open_dataarray(PATH_PP+'Bm_x_MLD_lowres.nc').isel(lon=slice(0,-3))
Bm_y_MLD_lowres = xr.open_dataarray(PATH_PP+'Bm_y_MLD_lowres.nc').isel(lon=slice(0,-3))
Bm_z_MLD_lowres = xr.open_dataarray(PATH_PP+'Bm_z_MLD_lowres.nc').isel(lon=slice(0,-3))



# combine datasets
grad_b = xr.concat([Bm_x_MLD_lowres, Bm_y_MLD_lowres, Bm_z_MLD_lowres],"axis")
fluxes = xr.concat([UsBs_MLD_lowres, VsBs_MLD_lowres, WsBs_MLD_lowres],"axis")

F_total_mag = np.sqrt(fluxes[0]*fluxes[0]+fluxes[1]*fluxes[1]+ fluxes[2]*fluxes[2])
```

# Skew and Residual flux decomposition

## Residual flux

```python
# residual flux
F_dot_grad_b = fluxes[0]*grad_b[0]+fluxes[1]*grad_b[1]+ fluxes[2]*grad_b[2]
grad_b_dot_grad_b = grad_b[0]*grad_b[0]+grad_b[1]*grad_b[1]+ grad_b[2]*grad_b[2]

F_res=  ((F_dot_grad_b * grad_b)/grad_b_dot_grad_b).transpose("axis", "time", "lat","lon")

F_res_mag = np.sqrt(F_res[0]*F_res[0]+F_res[1]*F_res[1]+ F_res[2]*F_res[2])
```

## Skew flux

```python
# skew flux
F_cross_grad_b= xr.DataArray(data=[fluxes[1]*grad_b[2]-fluxes[2]*grad_b[1], fluxes[2]*grad_b[0]-fluxes[0]*grad_b[2],
                           fluxes[0]*grad_b[1]-fluxes[1]*grad_b[0]],
                     dims=["axis", "time", "lat","lon"],
                     coords=dict(
                         axis=[0,1,2],
                         lon=fluxes.lon,
                         lat=fluxes.lat,
                         time=fluxes.time,
    ),)


F_cross_grad_b_cross_grad_b= xr.DataArray(data=[F_cross_grad_b[1]*grad_b[2]-F_cross_grad_b[2]*grad_b[1], 
                                                F_cross_grad_b[2]*grad_b[0]-F_cross_grad_b[0]*grad_b[2],
                                                F_cross_grad_b[0]*grad_b[1]-F_cross_grad_b[1]*grad_b[0]],
                     dims=["axis", "time", "lat","lon"],
                     coords=dict(
                         axis=[0,1,2],
                         lon=fluxes.lon,
                         lat=fluxes.lat,
                         time=fluxes.time,
    ),)


F_skew = F_cross_grad_b_cross_grad_b/grad_b_dot_grad_b

F_skew_mag = np.sqrt(F_skew[0]*F_skew[0]+F_skew[1]*F_skew[1]+ F_skew[2]*F_skew[2])

```

# Visualize

F_skew = $$|Flux \times \grad b|/ |\grad b|$$

F_res = $$|Flux \cdot \grad b|/ |\grad b|$$

Flux is total flux vector

```python
# visualize
it=160
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 3))
fig.tight_layout() # Or equivalently,  "plt.tight_layout()"

plt.subplot(131)
F_total_mag.isel(time=it).plot(vmin=0,vmax=1.3e-5,cmap='viridis')
plt.title('total')

plt.subplot(132)
F_skew_mag.isel(time=it).plot(vmin=0,vmax=1.3e-5,cmap='viridis')
plt.title('skew')


plt.subplot(133)
F_res_mag.isel(time=it).plot(vmin=0,vmax=1.3e-5,cmap='viridis')
plt.title('residual')
```
