```python
# Load all datasets
PATH_PP = PATH+'preprocessed_data/'

# mesoscale fields
Bm_MLD_lowres = xr.open_dataarray(PATH_PP+'Bm_MLD_lowres.nc')
Um_MLD_lowres = xr.open_dataarray(PATH_PP+'Um_MLD_lowres.nc')
Vm_MLD_lowres = xr.open_dataarray(PATH_PP+'Vm_MLD_lowres.nc')
Wm_MLD_lowres = xr.open_dataarray(PATH_PP+'Wm_MLD_lowres.nc')

# submesoscale fields
UsBs_MLD_lowres = xr.open_dataarray(PATH_PP+'UsBs_MLD_lowres.nc')
VsBs_MLD_lowres = xr.open_dataarray(PATH_PP+'VsBs_MLD_lowres.nc')
WsBs_MLD_lowres = xr.open_dataarray(PATH_PP+'WsBs_MLD_lowres.nc')

# buoyancy gradients
Bm_x_MLD_lowres = xr.open_dataarray(PATH_PP+'Bm_x_MLD_lowres.nc')
Bm_y_MLD_lowres = xr.open_dataarray(PATH_PP+'Bm_y_MLD_lowres.nc')
Bm_z_MLD_lowres = xr.open_dataarray(PATH_PP+'Bm_z_MLD_lowres.nc')
```

```python
# spatial correlation
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

plt.subplot(221)
Bm_WsBs_corr = xr.corr(WsBs_MLD_lowres, Bm_MLD_lowres,dim='time')
Bm_WsBs_corr.plot()
plt.title('Bm WsBs Correlation')

plt.subplot(222)
Um_WsBs_corr = xr.corr(WsBs_MLD_lowres, Um_MLD_lowres,dim='time')
Um_WsBs_corr.plot()
plt.title('Um WsBs Correlation')


plt.subplot(223)
Vm_WsBs_corr = xr.corr(WsBs_MLD_lowres, Vm_MLD_lowres,dim='time')
Vm_WsBs_corr.plot()
plt.title('Vm WsBs Correlation')


plt.subplot(224)
Wm_WsBs_corr = xr.corr(WsBs_MLD_lowres, Wm_MLD_lowres,dim='time')
Wm_WsBs_corr.plot()
plt.title('Wm WsBs Correlation')
```

```python
# temporal correlation
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

plt.subplot(221)
Bm_WsBs_corr = xr.corr(WsBs_MLD_lowres, Bm_MLD_lowres,dim=['lon','lat'])
Bm_WsBs_corr.plot()
plt.title('Bm WsBs Correlation')

plt.subplot(222)
Um_WsBs_corr = xr.corr(WsBs_MLD_lowres, Um_MLD_lowres,dim=['lon','lat'])
Um_WsBs_corr.plot()
plt.title('Um WsBs Correlation')


plt.subplot(223)
Vm_WsBs_corr = xr.corr(WsBs_MLD_lowres, Vm_MLD_lowres,dim=['lon','lat'])
Vm_WsBs_corr.plot()
plt.title('Vm WsBs Correlation')


plt.subplot(224)
Wm_WsBs_corr = xr.corr(WsBs_MLD_lowres, Wm_MLD_lowres,dim=['lon','lat'])
Wm_WsBs_corr.plot()
plt.title('Wm WsBs Correlation')
```

```python
# spatial correlation
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 3))

plt.subplot(131)
Bm_x_WsBs_corr = xr.corr(WsBs_MLD_lowres, Bm_x_MLD_lowres,dim='time')
Bm_x_WsBs_corr.plot()
plt.title('Bm_x WsBs Correlation')

plt.subplot(132)
Bm_y_WsBs_corr = xr.corr(WsBs_MLD_lowres, Bm_y_MLD_lowres,dim='time')
Bm_y_WsBs_corr.plot()
plt.title('Bm_y WsBs Correlation')


plt.subplot(133)
Bm_z_WsBs_corr = xr.corr(WsBs_MLD_lowres, Bm_z_MLD_lowres,dim='time')
Bm_z_WsBs_corr.plot()
plt.title('Bm_z WsBs Correlation')

```

```python
# average correlation
Bm_WsBs_corr = xr.corr(WsBs_MLD_lowres, Bm_MLD_lowres).values
Um_WsBs_corr = xr.corr(WsBs_MLD_lowres, Um_MLD_lowres).values
Vm_WsBs_corr = xr.corr(WsBs_MLD_lowres, Vm_MLD_lowres).values
Wm_WsBs_corr = xr.corr(WsBs_MLD_lowres, Wm_MLD_lowres).values

Bm_x_WsBs_corr = xr.corr(WsBs_MLD_lowres, Bm_x_MLD_lowres).values
Bm_y_WsBs_corr = xr.corr(WsBs_MLD_lowres, Bm_y_MLD_lowres).values
Bm_z_WsBs_corr = xr.corr(WsBs_MLD_lowres, Bm_z_MLD_lowres).values

corr_array = np.array([Bm_WsBs_corr, Um_WsBs_corr, Vm_WsBs_corr, Wm_WsBs_corr, Bm_x_WsBs_corr,
              Bm_y_WsBs_corr, Bm_z_WsBs_corr])
corr_vals = np.arange(-1,1,0.1)

corr_array_pos = np.array([Bm_WsBs_corr, Um_WsBs_corr, Vm_WsBs_corr, Wm_WsBs_corr, Bm_x_WsBs_corr,
              Bm_y_WsBs_corr, Bm_z_WsBs_corr])
corr_array_pos[corr_array<0] = np.nan
corr_array_neg = np.array([Bm_WsBs_corr, Um_WsBs_corr, Vm_WsBs_corr, Wm_WsBs_corr, Bm_x_WsBs_corr,
              Bm_y_WsBs_corr, Bm_z_WsBs_corr])
corr_array_neg[corr_array>0] = np.nan
```
